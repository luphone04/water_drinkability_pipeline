### **SPLITTING DATASET**

```python
from sklearn.model_selection import train_test_split
```

```python
train_data, test_data = train_test_split(data, test_size=0.20, random_state=42)
train_data.head()
```

```python
test_data.head()
```

```python
def fill_missing_with_media(df):
  for column in df.columns:
    if df[column].isnull().any():
      median_value = df[column].median()
      df[column].fillna(median_value, inplace=True)
  return df
```

```python
train_processed_data = fill_missing_with_media(train_data)
test_processed_data = fill_missing_with_media(test_data)
```

```python
from sklearn.ensemble import RandomForestClassifier
import pickle

X_train = train_processed_data.iloc[:,0:-1].values
y_train = train_processed_data.iloc[:,-1].values

clf = RandomForestClassifier()
clf.fit(X_train,y_train)
```

```python
pickle.dump(clf,open("/content/model.pkl","wb"))
```

```python
X_test = test_processed_data.iloc[:,0:-1].values
y_test = test_processed_data.iloc[:,-1].values
```

```python
from sklearn.metrics import accuracy_score,precision_score, recall_score, f1_score
model = pickle.load(open('/content/model.pkl',"rb"))
y_pred = model.predict(X_test)
acc = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1_score = f1_score(y_test,y_pred)
```

```python
print(acc)
print(precision)
print(recall)
print(f1_score)
```
